```python
from camera_placement.multiview_metaworld import CameraConfig
from camera_placement.multiview_metaworld import MultiViewMetaworldEnv, ENV_POLICY_MAP
import gymnasium as gym
import time 
from copy import deepcopy

```

```python
from camera_placement.metaworld_viewpoints import METAWORLD_CAMERA_VIEWPOINT_CONFIGS

all_configs = METAWORLD_CAMERA_VIEWPOINT_CONFIGS
```

```python
env_name = "handle-pull-v3"
viewpoint_dict = {}

for i, camera_config in enumerate(all_configs):
    env = gym.make("Meta-World/multiview", metaworld_env_name=env_name, cameras_config=[camera_config])
    obs, info = env.reset()
    img = obs["pixels"][camera_config.uid]
    viewpoint_dict[f"viewpoint_az={camera_config.azimuth}_el={camera_config.elevation}_dist={camera_config.distance}"] = img



```

```python
# display all viewpoints in a grid
import matplotlib.pyplot as plt

fig, axes = plt.subplots(len(viewpoint_dict)//5+1, 5, figsize=(15,5*len(viewpoint_dict)//5+1))

for i, viewpoint in enumerate(viewpoint_dict):
    col,row = i//5, i%5
    axes[col,row].imshow(viewpoint_dict[viewpoint])
    axes[col,row].axis('off')
```
